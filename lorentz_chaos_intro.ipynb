```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from math import *
```

```python
# plot formatting
plt.rc("text", usetex=False) # renders LaTeX more quickly
plt.rc("font", family = "serif",size=14) # font specifications
plt.rc("figure",figsize=(14,12)) # figure size
%config InlineBackend.figure_format = 'retina' # retina-display quality
```

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
dxdt = np.sin(x)
```

```python
#plot x y
plt.plot(x, dxdt, linewidth=3)
plt.title("Sin(x)")
plt.show()
```

```python
def lorenz(t, r, rho, sigma, beta):
    x,y,z = r
    dxdt = sigma*(y-x)
    dydt = x*(rho-z) - y
    dzdt = x*y-beta*z
    return np.array([dxdt, dydt, dzdt])

lorenz_init = np.array([1,1,1])
```

```python
#use ρ = 28, σ = 10, and β = 8/3
lorenz_sol = solve_ivp(lorenz, (0,100), lorenz_init, args=(28,10,8./3), dense_output=True)
```

```python
t_eval = np.linspace(0, 100, 10000)
```

```python
r = lorenz_sol.sol(t_eval)
```

```python
#plot x y
plt.plot(r[0], r[1], linewidth=1)
plt.title("Lorenz Attractor XY")
plt.show()
```

```python
#plot y z
plt.plot(r[1], r[2], linewidth=1)
plt.title("Lorenz Attractor YZ")
plt.show()
```

```python
# I google "matplotlib 3d plot example" every single time I do this
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
fig.suptitle(r"Lorenz Attractor", fontsize=20)
ax.plot(r[0], r[1], r[2], linewidth=1,c='xkcd:purple')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
